### ANALYZING FREQUENCY OF OSCILLATIONS IN FIGURE 7 (SM)

```python
times_short_time = np.array(times_short_time)                 
values_short_time = np.array(exp_values_nθ12_UP_short_time) 
dt_short_time = times_short_time[1] - times_short_time[0]
N_short_time = len(times_short_time)

# ---------------- FFT obtention, dominant frequency and period ----------------
values_centered_short_time = values_short_time - np.mean(values_short_time)
fft_vals_short_time = np.fft.rfft(values_centered_short_time)
freqs_short_time = np.fft.rfftfreq(N_short_time, d=dt_short_time)           
fft_amp_short_time = 2.0 * np.abs(fft_vals_short_time) / N_short_time

peak_index_short_time = np.argmax(fft_amp_short_time[1:]) + 1  
f_peak_short_time = freqs_short_time[peak_index_short_time]             
omega_peak_short_time = 2 * np.pi * f_peak_short_time        
T_peak_short_time = 1.0 / f_peak_short_time                 

print("========== FFT RESULTS ==========")
print(f"Peak frequency f      = {f_peak_short_time:.6e} = {f_peak_short_time}  (cycles / time)")
print(f"Angular frequency ω   = {omega_peak_short_time:.6e} = {omega_peak_short_time} (radians / time)")
print(f"Period T              = {T_peak_short_time:.6e} = {T_peak_short_time} (time)")

# ---------------- Visually-obtained frequency and period ----------------
T_visual_short_time = 2587.0  
f_visual_short_time = 1.0 / T_visual_short_time
omega_visual_short_time = 2 * np.pi * f_visual_short_time

print("\n========== VISUAL RESULTS ==========")
print(f"Peak frequency f      = {f_visual_short_time:.6e} = {f_visual_short_time}  (cycles / time)")
print(f"Angular frequency ω   = {omega_visual_short_time:.6e} = {omega_visual_short_time} (radians / time)")
print(f"Period T              = {T_visual_short_time:.6e} = {T_visual_short_time} (time)")

# ---------------- Perturbation theory frequency and period ----------------
omega_theory_short_time = 5 * λ_short_time**4 / (16 * m_short_time**3)

print("\n========== THEORY ==========")
print(
    f"Theoretical f         = {omega_theory_short_time / (2 * np.pi):.6e} = {omega_theory_short_time / (2 * np.pi)}"
    f" = (5*λ^4/(16*m^3))/(2*np.pi) (cycles / time)"
)
print(f"Theoretical ω         = {omega_theory_short_time:.6e} = {omega_theory_short_time} = 5*λ^4/(16*m^3)")
print(f"Period T_theory      = {2*np.pi/omega_theory_short_time:.3e} = {2*np.pi/omega_theory_short_time} (time)")

# ---------------- Relative errors ----------------
def relative_error(value, reference):
    return (value - reference) / reference

def print_error_block(title, f_val, omega_val, T_val, omega_theory):
    print(f"\n========== ERRORS {title} ==========")

    f_theory = omega_theory / (2 * np.pi)
    T_theory = 2 * np.pi / omega_theory

    err_f = relative_error(f_val, f_theory)
    err_omega = relative_error(omega_val, omega_theory)
    err_T = np.abs(relative_error(T_val, T_theory))

    print(f"Relative error f      = {err_f:.3e} = {err_f}")
    print(f"Relative error ω      = {err_omega:.3e} = {err_omega}")
    print(f"Relative error period = {err_T:.3e} = {err_T}")

print_error_block(
    "FFT",
    f_peak_short_time,
    omega_peak_short_time,
    T_peak_short_time,
    omega_theory_short_time
)

print_error_block(
    "VISUAL",
    f_visual_short_time,
    omega_visual_short_time,
    T_visual_short_time,
    omega_theory_short_time
)
```
