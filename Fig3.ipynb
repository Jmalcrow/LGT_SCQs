### This notebook reproduces the vortex dynamics obtained for different $\lambda/g$ and $\lambda/m$ ratios (50,100,300) for $N=13$ and $t_{\rm max} = 1000/100$ in "Data_Fig3.ipynb"

```python
λ_m = 50.0 #300#100#50.0     # λ/m
λ_g = 50.0 #300#100#50.0    # λ/g
N = 13
t_max = 1000.0

filename_vortex = (
    f"LGT_SCQs_plaquette_reduced_def_OCT25/"
    f"Vortex_dynamics_multiple_theta_mathieu_basis_"
    f"N_{int(N)}_tmax_{int(t_max)}_λ_m_{int(λ_m)}_λ_g_{int(λ_g)}.npz"
)

data_vortex = np.load(filename_vortex)
theta_list = data_vortex["thetas"]      
times      = data_vortex["t"]          
P_ops      = data_vortex["P_ops"]       
n_sqrs     = data_vortex["n_sqrs"]      
Fidelities = data_vortex["Fidelities"]  
```

```python
plt.rcParams.update({
    "figure.dpi": 100,
    "font.size": 24,
    "font.family": "serif",
    "text.usetex": True,
    "axes.labelsize": 36,
    "xtick.labelsize": 34,
    "ytick.labelsize": 34,
    "axes.titlesize": 14,
    "axes.grid": False,
    "lines.linewidth": 4,
    "lines.markersize": 6,
    "legend.fontsize": 14,
    "legend.title_fontsize": 16,
    "legend.frameon": False,
    "savefig.dpi": 150,
    "savefig.bbox": "tight",
})

n_theta = len(theta_list)

colors = plt.cm.viridis(np.linspace(0, 1, n_theta))

labels = [r"$\pi/4$", r"$\pi/2$", r"$3\pi/4$", r"$\pi$"]

fig, axs = plt.subplots(3, 1, figsize=(14, 11.1), sharex=True, dpi=300)

# ---------------- Plaquette operator ----------------
for i in range(n_theta):
    axs[0].plot(
        times, 
        P_ops[i], 
        color=colors[i], 
        label=labels[i]
)
axs[0].set_ylabel(r"$\langle \hat{\square} \rangle$", labelpad=10)
axs[0].grid(True)
axs[0].set_ylim(-1.02, 1.02)
axs[0].set_xlim(0, t_max)
axs[0].set_yticks([-1, 0, 1])
leg_b = axs[0].legend(
    loc='lower center',
    bbox_to_anchor=(0.55, 1.03),
    ncol=4,
    fontsize=32,
    frameon=False,
    handletextpad=0.4,
    columnspacing=1.0,
    title=None
)
axs[0].text(
    0.11, 1.33,
    r"$\mathbf{\Theta}$",
    transform=axs[0].transAxes,
    fontsize=32,
    va='center'
)
axs[0].text(-180.0, 0.9, r"\rm (a)", fontsize=34)

# ---------------- n_12^2 operator ----------------
for i in range(n_theta):
    axs[1].plot(
        times, 
        n_sqrs[i], 
        color=colors[i]
)
axs[1].set_ylabel(r"$\langle \hat{n}_{12}^2 \rangle$", labelpad=20)
axs[1].grid(True)
if λ_m == 50.0 and λ_g == 50.0:
    axs[1].set_ylim(0, 10.0) 
    axs[1].set_yticks([0, 8, 16]) 
elif λ_m == 100.0 and λ_g == 100.0:
    axs[1].set_ylim(0, 30.0) 
    axs[1].set_yticks([0, 15, 30]) 
elif λ_m == 300.0 and λ_g == 300.0:
    axs[1].set_ylim(0, 80.0) 
    axs[1].set_yticks([0, 40, 80]) 
axs[1].set_xlim(0, t_max)
axs[1].text(-180.0, 15, r"\rm (b)", fontsize=34)

# ---------------- Fidelity ----------------
for i in range(n_theta):
    axs[2].plot(
        times, 
        Fidelities[i], 
        color=colors[i]
)
axs[2].set_ylabel(
    r"$\mathcal{F}_{\Theta} = |\langle \Theta | \Psi_t \rangle|^2$",
    labelpad=10
)
axs[2].grid(True)
axs[2].set_ylim(0, 1)
axs[2].set_xlim(0, t_max)
axs[2].set_yticks([0, 0.5, 1.0])
axs[2].set_xlabel(r"$\lambda t$")
axs[2].text(-180.0, 0.95, r"\rm (c)", fontsize=34)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.savefig(
    f"LGT_SCQs_plaquette_reduced_def_OCT25/"
    f"Vortex_dynamics_multiple_theta_mathieu_basis_N_{int(N)}_"
    f"tmax_{int(t_max)}_λ_m_{int(λ_m)}_λ_g_{int(λ_g)}_long_time.pdf",
    dpi=600
)
plt.show()
```
